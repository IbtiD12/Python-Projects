The result of this project is an MLflow experiment with three different regression models (linear regression, decision tree regressor, and random forest regressor) and their results.

We will log the RMSE scores for each model as a metric called rmse_{{model}}, where {{model}} is replaced with lr, tr, and fr for linear regression, decision tree regressor, and random forest regressor respectively.

Store selected features in feature_selection and the target variable as target_var.

Define a function called preprocess_df(), which takes three arguments: df- the DataFrame for preprocessing; feature_selection - the features to subset the DataFrame on; and target_var - the target variable for the predictive model. 
Define a function predict_and_evaluate(), which takes three arguments: model - a pre-trained ML model; x_test and y_test - test set for independent variables and labels respectively. The function should predict values from the test set for the particular model, and calculate and return the root mean squared error (RSME).

Complete the for loop for varying depths, where I train the models, evaluate their performance using the predict_and_evaluate() function, and log their performance using the current depth.

```python
import pandas as pd
import numpy as np
import mlflow
import mlflow.sklearn
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# Load data and perform exploratory analysis
df = pd.read_csv("london_weather.csv")
#View the data
df.head()

#1. Basic information
df.info()
#2. Describe the data - descriptive statistics
df.describe()

#3. Find the duplicates
df.duplicated().sum()

#4. Find null values
df.isnull().sum()

#Converting for visulization
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
# 5. Replace null values with mean
df_metrics = ['cloud_cover', 'sunshine', 'global_radiation', 'max_temp', 'mean_temp', 'min_temp', 'precipitation', 'pressure', 'snow_depth']
df_per_month = df.groupby(['year', 'month'], as_index = False)[df_metrics].mean()

#6. Visulizing data
#6.1 corrilation
#Correlation 
df.corr()

#6.2 Lineplot
sns.lineplot(x="year", y="mean_temp", data=df_per_month, ci=None)
plt.show()

#6.3 barplot
sns.barplot(x='month', y='precipitation', data=df)
plt.show()

 #6.4 heatmap
sns.heatmap(df.corr())

feature_selection = ["month","cloud_cover","sunshine","global_radiation","precipitation","pressure"]
target_var = 'mean_temp'

df=df.dropna(subset=['mean_temp'])

def preprocess_df(df, feature_selection, target_var):
    """
    Split dataframe into X and y, and train and test consecutively. Then impute and scale both train and test features. Returns the train and test ets
    """
    X = df[feature_selection]    
    y = df[target_var]
    
    # Complete this function
    X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=10,test_size=0.33)
    
    #Impute Missing values
    imputer=SimpleImputer(strategy="mean")
    #fit on the training data
    X_train=imputer.fit_transform(X_train)
    X_test=imputer.transform(X_test)
    
    #Scale the dataset
    scaler=StandardScaler()
    #fit on the data
    X_train=scaler.fit_transform(X_train)
    X_test=scaler.transform(X_test)
    
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = preprocess_df(df, feature_selection, target_var)

def predict_and_evaluate(model, x_test, y_test):
    """
    Predict values from test set, calculate and return the root mean squared error.
    """
    # Complete this function
    y_pred=model.predict(x_test)
    rmse=np.sqrt(mean_squared_error(y_test,y_pred))
    return rmse

EXPERIMENT_NAME = "weather_predict"

try:
    EXPERIMENT_ID = mlflow.get_experiment_by_name(EXPERIMENT_NAME).experiment_id
    print(f"Using existing experiment with ID: {EXPERIMENT_ID}")
except MlflowException:
    # If the experiment doesn't exist, create a new one
    print(f"Creating a new experiment: {EXPERIMENT_NAME}")
    EXPERIMENT_ID = mlflow.create_experiment(EXPERIMENT_NAME)

# Adjust the parameters
max_depth_parameters = [1, 2]

for idx, depth in enumerate([1, 2, 5, 10, 20]):
    parameters = {
        'max_depth': depth,
        'random_state': 1
    }    
    RUN_NAME = f"run_{idx}"
    # Complete the experiment loop
    with mlflow.start_run(experiment_id=EXPERIMENT_ID,run_name=RUN_NAME):
        lr = LinearRegression().fit(X_train, y_train)
        fr = RandomForestRegressor(random_state = 42,max_depth=depth).fit(X_train, y_train)
        tr = DecisionTreeRegressor(random_state = 42,max_depth=depth).fit(X_train, y_train)
        mlflow.sklearn.log_model(lr,"lr")
        mlflow.sklearn.log_model(fr,"fr")
        mlflow.sklearn.log_model(tr,"tr")
        #evaluate performance
        lin_reg_rmse=predict_and_evaluate(lr,X_test, y_test)
        rf_reg_rsme=predict_and_evaluate(fr,X_test, y_test)
        dt_reg_rsme=predict_and_evaluate(tr,X_test, y_test)
    
         # Log performance
        mlflow.log_param("max_depth", depth)
        mlflow.log_metric("rmse_lr", lin_reg_rmse)
        mlflow.log_metric("rmse_tr", dt_reg_rsme)
        mlflow.log_metric("rmse_fr", rf_reg_rsme)
        
        
    

experiment_results = mlflow.search_runs(experiment_names=[EXPERIMENT_NAME])
experiment_results
```
